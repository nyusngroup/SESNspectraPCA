# Tests the write_lnw() function in SNIDsn.py for SN2004gt and compares it to the original SN2004gt SNID template 

```python
import SNIDsn
import SNIDdataset
import numpy as np
import matplotlib.pyplot as plt
```

```python
sn04gt = SNIDsn.SNIDsn()
```

```python
sn04gt.loadSNIDlnw('Tutorial_Data/sn2004gt.lnw')
```

#### Creates the new SNID template 'new_sn2004gt.lnw'

```python
sn04gt.write_lnw()
```

#### Initializes and loads in the new template

```python
sn04gt_new = SNIDsn.SNIDsn()
```

```python
sn04gt_new.loadSNIDlnw('new_sn2004gt.lnw')
```

#### Initializes lists of the original (old) data from 'sn2004gt.lnw' and the test (new) data from 'new_sn2004gt.lnw'

```python
old = sn04gt.data.tolist()
new = sn04gt_new.data.tolist()
```

#### Creates 13 pairs of subplots that compare each phase of the original and test data. The first pair of subplots compare the first phase, the second pair compares the second phase, etc.  

```python
for i in range(13):
    old_phase = []
    new_phase = []
    for j in range(1024):
        old_phase.append(old[j][i])
        new_phase.append(new[j][i])
    plt.subplot(2, 1, 1)
    plt.plot(sn04gt.wavelengths, old_phase)
    plt.title('Original vs New')
    plt.ylabel('Flux')
    
    plt.subplot(2, 1, 2)
    plt.plot(sn04gt_new.wavelengths, new_phase)
    plt.xlabel('Wavelength')
    plt.ylabel('Flux')
    plt.tight_layout()
    plt.show()
```

#### The pairs of subplots shown above are the same, so the original and test data are exactly the same
